##### Criando variáveis globais do projeto 

```python
A1,B1,C1 = "","",""
A2,B2,C2 = "","",""
A3,B3,C3 = "","",""
table_num = 0
```

### 2. Criando função que recebe comando de voz pt-BR

```python
import speech_recognition as sr
from IPython.display import clear_output

def get_voice_message():

    global A1
    global A2
    global A3
    global B1
    global B2
    global B3
    global C1
    global C2
    global C3
    global table_num

    r = sr.Recognizer()
    request = ""
    counter = 0
    while True:
        try:
            with sr.Microphone() as source:
                if counter >=5: 
                    clear_output()
                    counter = 0

                print("Ouvindo...")
                counter = counter + 1
                r.adjust_for_ambient_noise(source, duration=0.5)
                audio = r.listen(source)
                text = r.recognize_google(audio,language="pt-BR")
                text = text.lower()  
            
            if "enviar" in text:
                print("Enviando...")

                list_of_string = request.split()

                try:

                    result_Index = list_of_string.index("adicionar") + 1
                    colunm_Index = list_of_string.index("coluna") + 1
                    line_Index = list_of_string.index("linha") + 1

                    if list_of_string[colunm_Index] == '1':
                        if list_of_string[line_Index] == '1': A1 = list_of_string[result_Index]
                        elif list_of_string[line_Index] == '2': A2 = list_of_string[result_Index]
                        else: A3 = list_of_string[result_Index]
                    elif list_of_string[colunm_Index] == '2':
                        if list_of_string[line_Index] == '1': B1 = list_of_string[result_Index]
                        elif list_of_string[line_Index] == '2': B2 = list_of_string[result_Index]
                        else: B3 = list_of_string[result_Index]
                    else:
                        if list_of_string[line_Index] == '1': C1 = list_of_string[result_Index]
                        elif list_of_string[line_Index] == '2': C2 = list_of_string[result_Index]
                        else: C3 = list_of_string[result_Index]

                except Exception as e:
                    print(f"Error encontrado: {e}. Por favor siga as instruções para obter sucesso.")

                break
            elif "apagar" in text or "limpar" in text:
                print("Texto deletado com sucesso")
                request = ""
                counter = counter + 1
            elif "ver" in text or "visualizar" in text or "mostrar" in text:
                print("Você disse:", request)
                counter = counter + 1
            elif "fechar" in text or "sair" in text:
                print("Saida de voz finalizada")
                break
            else:
                request = request + text + " "
                print("Você disse:", request)
                counter = counter + 1
         
        except sr.RequestError as e:
            print("Erro na requisição, fale novamente; {0}".format(e))
            counter = counter + 1

        except sr.UnknownValueError:
            print("Não foi possível escutar o audio, fale novamente")
            counter = counter + 1

        except KeyboardInterrupt:
            print("Programa interrompido pelo usuário")
            break
    
    table_num = table_num + 1
```

##### Salvando audio ditado das instruções do projeto

```python
from gtts import gTTS

instructions = """ 
Para incluir valor às células da tabela diga 
"Adicionar (texto desejado) na coluna (números de 1 a 3) 
linha (números de 1 a 3)".
Para consultar o que foi dito diga "Ver" ou "Visualizar" 
ou "Mostrar". 
Para apagar o que foi dito diga "Limpar" ou "Apagar". 
Para enviar o que foi dito diga "Enviar". 
Caso queira interromper o áudio diga "Sair" ou "Fechar".
"""

gtts_object = gTTS(text=instructions, lang="pt-BR", slow=False)
instructions_audio = "instructions_audio.wav"

gtts_object.save(instructions_audio)
```

### 3. Criando interface do projeto

##### O projeto consiste em preencher células de uma determinada tabela utilizando comando de voz.

```python
import ipywidgets as widgets
from IPython.display import display, HTML,Audio,clear_output

def get_current_display(A1,B1,C1,A2,B2,C2,A3,B3,C3,table_num):

  global instructions_audio

  def on_button_click(b):
    with output:
        get_voice_message()
        #print("Python function executed!")

        global A1
        global A2
        global A3
        global B1
        global B2
        global B3
        global C1
        global C2
        global C3
        global table_num
        clear_output(wait=True)
        b.close()
        get_current_display(A1,B1,C1,A2,B2,C2,A3,B3,C3,table_num)
        

  # Create a button widget
  button = widgets.Button(description="Ativar microfone", button_style='primary')
  output = widgets.Output()

  html_instructions = """ 
  <h2>Instruções:</h2>
  <ol>
  <li>Para incluir valor às células da tabela diga <b>"Adicionar {texto desejado} na coluna {números de 1 a 3} linha {números de 1 a 3}"</b>.</li>
  <li>Para consultar o que foi dito diga <b>"Ver"</b> ou <b>"Visualizar"</b> ou <b>"Mostrar"</b>.</li>
  <li>Para apagar o que foi dito diga <b>"Limpar"</b> ou <b>"Apagar"</b>.</li>
  <li>Para enviar o que foi dito diga <b>"Enviar"</b>.</li>
  <li>Caso queira interromper o áudio diga <b>"Sair"</b> ou <b>"Fechar"</b>.</li>
  </ol>
  """

  html_table_code = f"""
  <br>
  <table>
    <tr>
      <th># {table_num}</th>
      <th>1</th>
      <th>2</th>
      <th>3</th>
    </tr>
    <tr>
      <td>1</td>
      <td>{A1}</td>
      <td>{B1}</td>
      <td>{C1}</td>
    </tr>
    <tr>
      <td>2</td>
      <td>{A2}</td>
      <td>{B2}</td>
      <td>{C2}</td>
    </tr>
    <tr>
      <td>3</td>
      <td>{A3}</td>
      <td>{B3}</td>
      <td>{C3}</td>
    </tr>
  </table>
  <br>
  <style>
    table, th, td {{
      border: 1px solid black;
      border-collapse: collapse;
      padding: 8px;
      text-align: center;
    }}
  </style>
  """
  # Link the button click event to the Python function
  button.on_click(on_button_click)

  # Display the button and the output area
  return display(HTML(html_instructions),Audio(instructions_audio), HTML(html_table_code),button,output)

def start():
  def on_button_click(b):
    with output:
        global A1
        global A2
        global A3
        global B1
        global B2
        global B3
        global C1
        global C2
        global C3
        global table_num
        b.close()
        get_current_display(A1,B1,C1,A2,B2,C2,A3,B3,C3,table_num)

  # Create a button widget
  button = widgets.Button(description="Clique para começar", button_style='danger')
  output = widgets.Output()

  button.on_click(on_button_click)

  return display(button,output)

start()
```
